```python
# Базовые статистика по количественным характеристикам
print('температура: 1. Нормальная 37.8, 2. Выше- инфекция, 3. Ниже - шок')
print('Значение моды: ', horse_task['rectal temperature'].round().mode()[0])
print(horse_task['rectal temperature'].describe(), end = '\n\n')
print('пульс: 1. Нормальная 30-40, 2. Выше - поражение 3. Ниже - редко у спортивных лошадей')
print('Значение моды: ', horse_task['pulse'].round().mode()[0])
print(horse_task['pulse'].describe(), end = '\n\n')
print('частота дыхания: Мало полезен в норме 8 - 10')
print('Значение моды: ', horse_task['respiratory rate'].round().mode()[0])
print(horse_task['respiratory rate'].describe(), end = '\n\n')
print('температура конечностей: 1. нормальные 2. Горячие- повышенная ректальные температура 3. Прохлданые 4. Холодные - шок')
print('Значение моды: ', horse_task['temperature of extremities'].round().mode()[0])
print(horse_task['temperature of extremities'].value_counts(), end = '\n\n')
print('возраст')
print(horse_task['Age'].value_counts(), end = '\n\n')
print('Боль')
print(horse_task['pain'].value_counts(), end = '\n\n')
print('Результат')
print(horse_task['outcome'].value_counts(), end = '\n\n')
```

```python
# Выбросы
print('температура: 1. Нормальная 37.8, 2. Выше- инфекция, 3. Ниже - шок')
tem_req_q1 = horse_task['rectal temperature'].quantile(0.25)
tem_req_q3 = horse_task['rectal temperature'].quantile(0.75)
tem_req_iqr = tem_req_q3 - tem_req_q1
tem_req_lower_bound = tem_req_q1 - (1.5 * tem_req_iqr) 
tem_req_upper_bound = tem_req_q3 + (1.5 * tem_req_iqr)
tem_req_remove_outliers = horse_task[horse_task['rectal temperature'].between(tem_req_lower_bound, tem_req_upper_bound, inclusive=True)].sort_values('rectal temperature')
pd.concat([horse_task, tem_req_remove_outliers]).drop_duplicates(keep=False).dropna()
```

```python
print('пульс: 1. Нормальная 30-40, 2. Выше - поражение 3. Ниже - редко у спортивных лошадей')
puls_q1 = horse_task['pulse'].quantile(0.25)
puls_q3 = horse_task['pulse'].quantile(0.75)
puls_iqr = puls_q3 - puls_q1
puls_lower_bound = puls_q1 - (1.5 * puls_iqr) 
puls_upper_bound = puls_q3 + (1.5 * puls_iqr)
puls_remove_outliers = horse_task[horse_task['pulse'].between(puls_lower_bound, puls_upper_bound, inclusive=True)].sort_values('pulse')
pd.concat([horse_task, puls_remove_outliers]).drop_duplicates(keep=False).dropna()
```

```python
print('частота дыхания: Мало полезен в норме 8 - 10')
res_rate_q1 = horse_task['respiratory rate'].quantile(0.25)
res_rate_q3 = horse_task['respiratory rate'].quantile(0.75)
res_rate_iqr = res_rate_q3 - res_rate_q1
res_rate_lower_bound = res_rate_q1 - (1.5 * res_rate_iqr) 
res_rate_upper_bound = res_rate_q3 + (1.5 * res_rate_iqr)
res_rate_remove_outliers = horse_task[horse_task['respiratory rate'].between(res_rate_lower_bound, res_rate_upper_bound, inclusive=True)].sort_values('respiratory rate')
pd.concat([horse_task, res_rate_remove_outliers]).drop_duplicates(keep=False).dropna()
```

# Столбец 'surgery'

```python
print(horse_task.groupby('outcome')['surgery?'].value_counts())
# Так как большинство успленных лощадей были взрослыми, то заполняем Nan==1
horse_task['surgery?'].fillna(horse_task['surgery?'].min(), inplace=True)
print(horse_task[horse_task['surgery?'].isna()])
```

# Столбец 'outcome'

```python
print(horse_task.groupby(['pain','temperature of extremities'])['outcome'].value_counts())
# Исходя из данных большиносто лошадей с показанием боли 4 и 3 степени температуры конечностей были усыплены, поэтому заполняем NaN == 3
horse_task['outcome'].fillna(horse_task['outcome'].max(), inplace=True)
```

# Столбец 'rectal temperature'

```python
horse_task[horse_task['rectal temperature'].isna()]
print(horse_task['rectal temperature'].mode()[0])
print(horse_task.groupby(['surgery?','outcome'])['rectal temperature'].describe())
#Так как разброс значении температуры незначительный (std), то заполняем Nan средним значением в зависимотси от наличия операции и результата
horse_task['rectal temperature'].fillna(horse_task.groupby(['surgery?','outcome'])['rectal temperature'].transform('mean'), inplace=True)
```

# Столбец pulse

```python
def cat_temp_req(tem):
    '''1- шок, 2- норма 3- инфекция'''
    if tem <= 37.7:
        return '1'
    elif 37.8 == tem:
        return '2'
    elif tem > 37.8:
        return '3'
#Разнесем ректальную температуру по категориям
horse_task['cat_temp']= horse_task['rectal temperature'].apply(cat_temp_req)
# Заменим Nan сердним исход из категории температуры 
#horse_task.groupby('cat_temp')['pulse'].describe()
horse_task['pulse'].fillna(horse_task.groupby(['cat_temp'])['pulse'].transform('mean'), inplace=True)
horse_task[horse_task['pulse'].isna()]
```

# Столбец respiratory rate

```python
#Так как полезность данного параметра сомнительна то заполним средним в завиисмоти от категории температуры


horse_task['respiratory rate'].describe()
horse_task['respiratory rate'].fillna(horse_task.groupby(['cat_temp'])['respiratory rate'].transform('mean'), inplace=True)
horse_task[horse_task['respiratory rate'].isna()]
```

# Столбец temperature of extremities

```python
# Заполним Nan в зависимости от ректальной температуры
def cat_temp_ext(tem):
    if 35.4 <= tem <= 36.5:
        return '4'
    elif 36.5< tem <=37.5:
        return '3'
    elif 37.5< tem <=38.0:
        return '1'
    elif tem > 38.0:
        return '2'
horse_task['temperature of extremities'].fillna(horse_task['rectal temperature'].apply(cat_temp_ext), inplace=True)
horse_task[horse_task['temperature of extremities'].isna()]
```

# Столбец pain

```python
horse_task[horse_task['pain'].isna()]
# Заполним столбец с покателем боли в зависимости от кочненого результата и данных о операции
'''
Наличие операции:
    1 = Да, была операция
    2 = Лечение прошло без хирургического вмешательства
Конечнй резуьтат
    1 = жил
    2 = умер
    3 = был усыплен
'''
def pain(surg, outcome):
    if surg == 1.0 and outcome == 3.0:
        return 4.0
    elif surg == 1.0 and outcome == 2.0:
        return 5.0
    elif surg == 1.0 and outcome == 1.0:
        return 1.0
    elif surg == 2.0 and outcome == 1.0:
        return 2.0
    elif  surg == 2.0 and outcome == 2.0:
        return 3.0
    elif  surg == 2.0 and outcome == 3.0:
        return 3.0
horse_task['pain'].fillna(horse_task.apply(lambda x: pain(x['surgery?'], x['outcome']), axis=1), inplace=True)
```
